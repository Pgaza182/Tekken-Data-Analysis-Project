```python
# load Libraries

import numpy as np
import pandas as pd
import matplotlib.pylab as plt
import matplotlib
```

```python
evo = pd.read_csv("Tekken_EVO_Results_2005-2024.csv")
print(evo)
```

Dataset only goes up to 2024 manually added EVO tournaments from 2025-26

```python
# Data Cleaning 
# Take out unecessary Columns that arent needed
# Format, P1/P2 Team

evo_clean = evo.drop(columns = ['Format', 'P1 Team', 'P2 Team'])
print(evo_clean.columns)
```

```python
# Divide the data set into 2 categories for the modern era (7, 8)

t7 = evo_clean[evo_clean['Game'] == 'Tekken 7']
t8 = evo_clean[evo_clean['Game'] == 'Tekken 8']

print( "Tekken 7 entries:", t7)
print( "Tekken 8 entries:", t8)
```

```python
# Combine P1 and P2 into a player list, drop the duplicate player names, then count the countries
t7_countries = pd.concat([
    t7[['P1', 'P1 Country']].rename(columns={'P1': 'Player', 'P1 Country': 'Country'}),
    t7[['P2', 'P2 Country']].rename(columns={'P2': 'Player', 'P2 Country': 'Country'})
]).drop_duplicates(subset='Player')['Country'].value_counts()

t8_countries = pd.concat([
    t8[['P1', 'P1 Country']].rename(columns={'P1': 'Player', 'P1 Country': 'Country'}),
    t8[['P2', 'P2 Country']].rename(columns={'P2': 'Player', 'P2 Country': 'Country'})
]).drop_duplicates(subset='Player')['Country'].value_counts()

# Plot Tekken 7 
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

t7_countries.plot(kind='bar', ax=axes[0], color='steelblue')
axes[0].set_title('Tekken 7 - Country Representation')
axes[0].set_xlabel('Country')
axes[0].set_ylabel('Appearances')

t8_countries.plot(kind='bar', ax=axes[1], color='crimson')
axes[1].set_title('Tekken 8 - Country Representation')
axes[1].set_xlabel('Country')
axes[1].set_ylabel('Appearances')

# Add numbers above bars
for ax in axes:
    for bar in ax.patches:
        ax.text(
            bar.get_x() + bar.get_width() / 2,
            bar.get_height() + 0.1,
            int(bar.get_height()),
            ha='center', va='bottom', fontsize=9
        )
```

We can see here the country distribution through the lifespan of Tekken 7 and Tekken 8 at EVO. Tekken 7 Obviously has more entrants because the game has been out for longer meaning more tournaments and more players. Korea has dominated both games with a dip in JP players placing top 8 in Tekken 8.

Now this is how the countries compare overall with just the count of players placing top 8. Lets look at how far each of these countries place in these top 8s. We will look at this by assigning numbers with higher numbers associating to lower placement and vice versa. Note that there can be data skewed as many KR can place just in top8-6 skewing their numbers towards less higher placings. 

```python
def get_tournament_placements(df):
    tournaments = df['EVO Year + Series'].unique()
    records = []
    
    for tournament in tournaments:
        t = df[df['EVO Year + Series'] == tournament].copy()
        
        # Handle Grand Finals reset - prioritize the Reset row over the normal one
        gf_rows = t[t['Tournament Round'] == 'Grand Finals']
        if 'Reset' in gf_rows['Special Condition'].values:
            gf = gf_rows[gf_rows['Special Condition'] == 'Reset'].iloc[0]
        else:
            gf = gf_rows.iloc[-1]
        
        winner = gf['Winner']
        gf_loser = gf['P1'] if gf['Winner'] == gf['P2'] else gf['P2']
        
        # Get all players
        p1s = t[['P1', 'P1 Country']].rename(columns={'P1': 'Player', 'P1 Country': 'Country'})
        p2s = t[['P2', 'P2 Country']].rename(columns={'P2': 'Player', 'P2 Country': 'Country'})
        all_players = pd.concat([p1s, p2s]).drop_duplicates(subset='Player')
        
        # Track FINAL elimination round for each player
        # Only track losses in LOSERS BRACKET rounds + Grand Finals
        losers_rounds = ['Losers R1', 'Losers R2', 'Losers R3', 'Losers Finals', 'Grand Finals']
        
        final_elimination = {}
        for _, row in t.iterrows():
            loser = row['P1'] if row['Winner'] == row['P2'] else row['P2']
            round_name = row['Tournament Round']
            
            if round_name in losers_rounds:
                # Always overwrite - last losers bracket loss is the true elimination
                final_elimination[loser] = round_name
        
        # Placement logic per your spec
        def get_placement(player):
            if player == winner:
                return 1
            if player == gf_loser:
                return 2
            elim_round = final_elimination.get(player)
            if elim_round == 'Losers Finals':
                return 3
            if elim_round == 'Losers R3':
                return 4
            if elim_round == 'Losers R2':
                return 5.5  # represents 5/6
            if elim_round == 'Losers R1':
                return 7.5  # represents 7/8
            return 7.5  # fallback for winners side semis losers etc
        
        for _, row in all_players.iterrows():
            player = row['Player']
            country = row['Country']
            records.append({
                'Tournament': tournament,
                'Player': player,
                'Country': country,
                'Placement': get_placement(player)
            })
    
    return pd.DataFrame(records)

t7_placements = get_tournament_placements(t7)
t8_placements = get_tournament_placements(t8)

placement_labels = {
    1: '1st',
    2: '2nd',
    3: '3rd',
    4: '4th',
    5.5: '5th/6th',
    7.5: '7th/8th'
}

pk = t7_placements[t7_placements['Country'] == 'PK'].copy()
pk['Placement'] = pk['Placement'].map(placement_labels)
print(pk)

print()

kr = t7_placements[t7_placements['Country'] == 'KR'].copy()
kr['Placement'] = kr['Placement'].map(placement_labels)
print(kr)
```

We can individually see each placement of players from PK and KR here in what tournmant they played in only for Tekken 7 Lets graph this out to get the average placements of both KR and PK for tekken 7.

```python
# Get average placement per country for T7
pk_avg7 = t7_placements[t7_placements['Country'] == 'PK']['Placement'].mean()
kr_avg7 = t7_placements[t7_placements['Country'] == 'KR']['Placement'].mean()

countries = ['PK', 'KR']
averages = [pk_avg7, kr_avg7]
colors = ['#01411C', '#C60C30']

fig, ax = plt.subplots(figsize=(7, 5))
bars = ax.bar(countries, averages, color=colors, width=0.4)

for bar, avg in zip(bars, averages):
    ax.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height() + 0.05,
        f'{avg:.2f}',
        ha='center', va='bottom', fontsize=11, fontweight='bold'
    )

ax.set_title('Tekken 7 EVO - Average Placement (PK vs KR)', fontsize=13)
ax.set_ylabel('Average Placement (lower = better)')
ax.set_ylim(0, max(averages) + 1)

for y, label in [(1, '1st'), (2, '2nd'), (3, '3rd'), (4, '4th')]:
    ax.axhline(y=y, color='gray', linestyle='--', alpha=0.4)
    ax.text(1.55, y + 0.05, label, fontsize=8, color='gray')

plt.tight_layout()
plt.show()

print(f'PK Average Placement: {pk_avg7:.2f} (across {len(t7_placements[t7_placements["Country"] == "PK"])} tournament entries)')
print(f'KR Average Placement: {kr_avg7:.2f} (across {len(t7_placements[t7_placements["Country"] == "KR"])} tournament entries)')
```

Now we got the big major influential countries lets do this for all the countries now.

```python
# Get average placement for all countries in T7
t7_country_avg = t7_placements.groupby('Country')['Placement'].agg(['mean', 'count']).reset_index()
t7_country_avg.columns = ['Country', 'Avg Placement', 'Entries']
t7_country_avg = t7_country_avg.sort_values('Avg Placement')

fig, ax = plt.subplots(figsize=(12, 5))
bars = ax.bar(t7_country_avg['Country'], t7_country_avg['Avg Placement'], color='steelblue')

# Add value + entry count above bars
for bar, (_, row) in zip(bars, t7_country_avg.iterrows()):
    label = f'{row["Avg Placement"]:.2f}'
    if row['Entries'] == 1:
        label += '*'
    ax.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height() + 0.05,
        label,
        ha='center', va='bottom', fontsize=9, fontweight='bold'
    )

for y, label in [(1, '1st'), (2, '2nd'), (3, '3rd'), (4, '4th')]:
    ax.axhline(y=y, color='gray', linestyle='--', alpha=0.4)
    ax.text(len(t7_country_avg) - 0.5, y + 0.05, label, fontsize=8, color='gray')

ax.set_title('Tekken 7 EVO - Average Placement by Country (lower = better)', fontsize=13)
ax.set_ylabel('Average Placement')
ax.set_xlabel('Country')
fig.text(0.5, -0.02, '* = only 1 tournament entry, may not be representative', 
         ha='center', fontsize=9, style='italic', color='gray')

plt.tight_layout()
plt.show()

print(t7_country_avg.to_string(index=False))
```
